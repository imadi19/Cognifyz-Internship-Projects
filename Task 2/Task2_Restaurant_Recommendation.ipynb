```python
# cell 1: Importing the libraries and loading dataset
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import StandardScaler

#let's Load datset
df = pd.read_csv("Dataset.csv")
df.head()
```

```python
# Cell 2: let's see basic Dataset info
df.info()
df.describe(include="all")
```

```python
# Cell 3: Handling missing values and cleaning dataset

# Check and display missing values before cleaning
print("Missing values before cleaning:")
print(df.isnull().sum())

# Fill missing values in the 'Cuisines' column with "Unknown"
df["Cuisines"] = df["Cuisines"].fillna("Unknown")

# Droping the columns that are irrelevant for analysis
columns_to_drop = [
    "Restaurant ID", "Address", "Locality Verbose", "Longitude", "Latitude",
    "Switch to order menu", "Rating color", "Rating text", "Country Code", "Currency"
]
df_cleaned = df.drop(columns=columns_to_drop)

# Check and display missing values after cleaning
print("\nMissing values after cleaning:")
print(df_cleaned.isnull().sum())

# Display the first few rows of the cleaned DataFrame
df_cleaned.head()
```

```python
encoder = LabelEncoder()

#Apply LabelEncoder to each column separately
df_cleaned["City"] = encoder.fit_transform(df_cleaned["City"])
df_cleaned["Locality"] = encoder.fit_transform(df_cleaned["Locality"])
df_cleaned["Cuisines"] = encoder.fit_transform(df_cleaned["Cuisines"])
df_cleaned["Has Table booking"] = encoder.fit_transform(df_cleaned["Has Table booking"])
df_cleaned["Has Online delivery"] = encoder.fit_transform(df_cleaned["Has Online delivery"])
df_cleaned["Is delivering now"] = encoder.fit_transform(df_cleaned["Is delivering now"])

df_cleaned.head()
```

```python
# Cell 5: Content-based filtering using cosine similarity
# Features to use for similarity
features = ["City", "Locality", "Cuisines", "Price range", "Average Cost for two", "Aggregate rating", "Votes"]
# Normalize features
scaler = StandardScaler()
X = scaler.fit_transform(df_cleaned[features])

# Compute similarity matrix
similarity = cosine_similarity(X)

# Show shape of similarity matrix and a sample
print("Similarity matrix shape:", similarity.shape)
print("\nSample similarity scores for first restaurant:\n", similarity[0][:10])
```

```python
# Cell 6: Function to recommend restaurants by name (content-based filtering)
def recommend_restaurant_by_name(name, n=5):
    # Find index of restaurant by name (first match)
    matches = df[df["Restaurant Name"].str.contains(name, case=False, na=False)]
    if matches.empty:
        return f"No restaurant found with name containing '{name}'"
    
    index = matches.index[0]
    
    # Get similarity scores for that restaurant
    scores = list(enumerate(similarity[index]))
    # Sort by similarity
    scores = sorted(scores, key=lambda x: x[1], reverse=True)
    # Get top n recommendations (excluding itself)
    top_indices = [i[0] for i in scores[1:n+1]]
    
    return df.iloc[top_indices][["Restaurant Name", "City", "Cuisines", "Average Cost for two", "Aggregate rating"]]

# Example usage
recommend_restaurant_by_name("Sambo Kojin", n=5)
```

```python
# Cell 7: Search restaurants by user preferences
def search_by_preferences(cuisine=None, price=None, rating=None, city=None, n=10):
    results = df.copy()
    
    if cuisine:
        results = results[results["Cuisines"].str.contains(cuisine, case=False, na=False)]
    if price:
        results = results[results["Price range"] == price]
    if rating:
        results = results[results["Aggregate rating"] >= rating]
    if city:
        results = results[results["City"].str.contains(city, case=False, na=False)]
    
    if results.empty:
        return "No restaurants found matching your preferences."
    
    return results[["Restaurant Name", "City", "Cuisines", "Average Cost for two", "Aggregate rating"]].head(n)

# Example usage: Indian restaurants with rating greater than or equal to 4
search_by_preferences(cuisine="Indian", rating=4, n=5)

```
